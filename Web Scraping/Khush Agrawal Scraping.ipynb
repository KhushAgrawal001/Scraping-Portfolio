```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://khushagrawal.tech')
```

```python
print(r)
```

```python
print(r.content)
```

```python
soup = BeautifulSoup(r.content , 'html.parser')
```

```python
print(soup.prettify)
```

```python
title = soup.title
print(title)
print(type(title))
print(type(title.string))
```

```python
div = soup.find('div', class_="about-desc")
print(div)
```

```python
about_desc_div = soup.find('div',class_="about-desc")
if about_desc_div:
    paragraph = about_desc_div.find('p')
    print(paragraph)
```

```python
if about_desc_div:
    bothparagraph = about_desc_div.find_all('p')
    print(bothparagraph)
```

```python
print(title.string)
for paragraph in bothparagraph:
    print(paragraph.get_text())

```
